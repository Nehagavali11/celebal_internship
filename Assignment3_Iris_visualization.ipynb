```python
# compare of sepal and petal width,len
x = df['SepalLengthCm']
y = df['SepalWidthCm']
plt.scatter(x,y,color = "green",marker = "+")

x = df['PetalLengthCm']
y = df['PetalWidthCm']
plt.scatter(x,y,color = "red",marker = ".")

plt.show()
```

```python
x = df['PetalLengthCm']
y = df['SepalLengthCm']
colors = np.random.choice(['hotpink', 'yellow'], size=len(x))
plt.scatter(x, y, c=colors,s = 80,marker = "*")
plt.title('Scatter plot of Petal Length vs Sepal Length')
plt.show()
```

```python
x = df['PetalWidthCm']
y = df['SepalWidthCm']
colors = np.random.choice(['gray', 'cyan'], size=len(x))
plt.scatter(x, y, c=colors)
plt.title('Scatter plot of Petal Width vs Sepal Width')
plt.show()
```

```python
# Count the occurrences of each species
species_counts = df['Species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.show()
```

```python
# Calculate the correlation matrix for the relevant numerical features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
corr_matrix = df[features].corr()

print(corr_matrix)

# Plot the heatmap for better visualization
sns.heatmap(corr_matrix, annot=True, cmap='cividis')
plt.title('Correlation Matrix of Iris Features')
plt.show()
```
